# EDA


## Using Seaborn


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'Dataset/tips.csv')
df.head()
```

### Correlation with Heatmap

- A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2d correlation matric (table) between two discrete dimension or event types. IT is very important in Feature selection


```python
df.info()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

# Join PLot

- A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a @D density plot


```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
# draws a regression line
```

## Pair Plot

A pair plot is also knows as a scatterplot, in which one variable in the same data row is matched with another variable's value.<br>
Pair plot are just elaborations on this, showing all variables paired with all other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")
```

```python
sns.pairplot(df, hue="smoker")
```

## Distplot

Helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'], )
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

## For Categorical values

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_csv('Dataset/titanic.csv')
```

```python
df.head()
```

## EDA

**Missing Values**


```python
df.isna().sum()
```

```python
sns.heatmap(df.isna(),yticklabels=False, cmap='viridis')
```

```python
sns.set_style('whitegrid')
sns.countplot(x ='Survived', data= df)
```

```python
sns.countplot(x ='Survived', data= df, hue='Sex')
```

```python
sns.countplot(x ='Survived', data= df, hue='Pclass', palette='rainbow')
```

```python
sns.displot(df['Age'].dropna(), kde = False, color='darkred', bins = 40)
```

```python
df['Age'].hist(bins=30, alpha = 0.3)
```

```python
sns.countplot(x='SibSp', data= df )
```
