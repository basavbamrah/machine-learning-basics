### Data Cleaning


```python
plt.figure(figsize=(12, 17))

sns.boxplot(x="Pclass", y="Age", data=df, palette="winter")
```

Here we get to know average age in each Pclass

- Now we can replace age values in each class based on above foind average


```python
def fill_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24

    return Age
```

```python
df["Age"] = df[["Age", "Pclass"]].apply(fill_age, axis=1)
```

```python
sns.heatmap(df.isna())
```

```python
# since Cabin has a lot of null values- Drop the column

df.drop(columns="Cabin", inplace=True)
df.head()
```

```python
sns.heatmap(df.isna())
```

## CAtegorigal Features


```python
df.head()
```

```python
df.info()
```

```python
sex = pd.get_dummies(df["Sex"], drop_first=True)
embark = pd.get_dummies(df["Embarked"], drop_first=True)

df.drop(columns=["Sex", "Embarked", "Name", "Ticket"], inplace=True)
df.head()
```

```python
df = pd.concat([df, sex, embark], axis=1)
df.head()
```

```python
y = df["Survived"]
X = df.drop(columns=["Survived"])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=200)
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
